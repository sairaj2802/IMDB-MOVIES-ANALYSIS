```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
data=pd.read_csv('/content/IMDB_Movies.csv')
```

```
data.head()
```

```
data.info()
```

Dropping Columns Which Is not necessary

```
data1=data.drop([
    'director_facebook_likes',
    'actor_3_facebook_likes',
    'actor_2_facebook_likes',
    'actor_1_facebook_likes',
    'cast_total_facebook_likes',
    'facenumber_in_poster',
    'plot_keywords',
    'aspect_ratio',
    'movie_facebook_likes'
    ],axis=1)
```

```
data1.info()
```

Handling Missing Values

Numeric Data


```
data1.isnull().sum()
```

```
row_missing_percentage=data1.isnull().mean(axis=1)*100
print(row_missing_percentage.sort_values(ascending=False).head(10))
```

```
data1=data1[row_missing_percentage<=30]
print(data1)
```

```
missing_by_column = data1.isnull().mean() * 100
print(missing_by_column.sort_values(ascending=False))
```

```
data1.isnull().sum()
```

```
data1['num_user_for_reviews'] = pd.to_numeric(data1['num_user_for_reviews'], errors='coerce')
```

```
data1['budget'] = data1['budget'].fillna(data1['budget'].median())
data1['gross'] = data1['gross'].fillna(data1['gross'].median())
data1['duration'] = data1['duration'].fillna(data1['duration'].median())
data1['num_critic_for_reviews'] = data1['num_critic_for_reviews'].fillna(data1['num_critic_for_reviews'].median())
data1['title_year'] = data1['title_year'].fillna(data1['title_year'].median())
data1['num_user_for_reviews'] = pd.to_numeric(data1['num_user_for_reviews'], errors='coerce')
data1['num_user_for_reviews'] = data1['num_user_for_reviews'].fillna(data1['num_user_for_reviews'].median())
```

Categorial Data


```
data1 = data1[data1['color'].notnull()]
```

```
data1 = data1[data1['director_name'].notnull()]
```

```
data1 = data1[data1['country'].notnull()]
```

```
data1 = data1[data1['language'].notnull()]
```

```
data1.to_csv('cleaned_IMDB_data.csv', index=False)
```
